**A neat feature - string indexing**

Using string indexing you can grab a portion of a string like so:

```python
print(message[4]) # this grabs the 5th character of the string

# why does this grab the 5th and not the 4th character?
# This is because programming languages start at index 0
```

image.png

**Explaining the syntax**:


String indexing can also be used to access a range of characters, known as string slicing. It follows the syntax `[start:end:step]`, where start is the starting index (inclusive), end is the ending index (exclusive), and step is the optional step value that specifies the increment between indices.

```python

print(message[0])           # output: "H"
print(message[-1])          # output: "o" (the last character in the string)
print(message[0:5])         # Output: "Hello"
print(message[7:])          # Output: "World!"
print(message[:5])          # Output: "Hello"
print(message[::2])         # Output: "Hlo ol!"
print(message[::-1])        # OUtput: "!dlroW olleH"
```

  - message[0] returns the first character
  - message[-1] returns the last character
  - message[0:5] returns the characters from index 0 to 4 (5 is exclusive), which is "Hello".
  - message[7:] returns the characters from index 7 to the end of the string, which is "World!".
  - message[:5] returns the characters from the beginning of the string to index 4 (5 is exclusive), which is "Hello".
  - message[::2] returns the characters with a step of 2, skipping every second character, which is "Hlo ol!".
  - message[::-1] returns the characters with a step of -1, starting at the end of the string and moving forward.

You can also use string indexing to change a string:

```python
sentence = "the dog barked"
print(sentence)

sentence = f"{sentence[:3]} cat {sentence[8:]}"
print(sentence)
```

**Lets break that down:**

I first set a variable for sentence as shown above.

Then I changed the value of the variable to be the string index of its previous value from characters 0 to 3 ("the") then added the string " cat " using string formatting then added the string index from characters 8 to the end of the string.

# *If Statement*

The if statement is the most fundamental conditional statement. It executes a block of code if a given condition is true. The basic syntax of an if statement is as follows:

```python
# if condition:
#     Code to be executed if the condition is true


x = 10

if x > 5:
    print("x is greater than 5")
```

 Notice how after the condition in the if statement there is a colon ":". This is important. This marks the end of the statement then after that the code continues on a new line and most importantly, the code is indented.

 This is the first time in the notebook where we see nested code. In Python, nesting code refers to placing blocks of code within other blocks of code. It allows you to create hierarchical structures and control the execution flow based on specific conditions. Indentation plays a crucial role in indicating the level of nesting and organizing the code.

# Nested Code

Nested code occurs when one code block is placed inside another code block. This can be seen when an if statement or a loop is present within another if statement or loop. The indentation level increases for each level of nesting.

Here's an example of nested code:

```python
'''
if condition1:
    # Outer if block
    statement1
    statement2
    if condition2:
        # Inner if block
        statement3
        statement4
'''
```

In this example, there is an outer if block and an inner if-else block. The inner block is nested within the outer block. The indentation visually represents the nested structure. The inner block is executed only if condition1 is true, and within the inner block, further decisions are made based on condition2.

Importance of Whitespace

Whitespace, specifically indentation, is significant in Python because it defines the structure and hierarchy of the code. It is used to indicate which statements belong to a specific code block and which statements are at a higher level of nesting. Python's use of indentation promotes code readability and helps in understanding the program's logic at a glance.

It's important to note that Python relies on consistent indentation to interpret code correctly. If the indentation is incorrect or inconsistent, it can lead to syntax errors or incorrect program behavior. Now back to logical conditionals.

# elif Statement

The elif statement allows you to check additional conditions if the preceding if statement or elif statement(s) (if any) are false. It is useful when you have multiple conditions to evaluate. The syntax is as follows:

```python
'''
if condition1:
    Code to be executed if condition1 is true
elif condition2:
    Code to be executed if condition1 is false and condition2 is true

'''
```

```python
x = 10

if x > 15:
    print("x is greater than 15")
elif x > 5:
    print("x is greater than 5 but not greater than 15")
```

In this example, the first condition x > 15 is evaluated, but it is false. Therefore, the program moves to the elif statement and checks the next condition x > 5. Since this condition is true, the corresponding code block is executed, and the message "x is greater than 5 but not greater than 15" is printed.

# else Statement

The else statement is used as a catch-all when all preceding conditions in an if-elif chain are false. It allows you to define a block of code to execute when none of the previous conditions evaluate to true. The syntax is as follows:

```python
x = 3

if x > 5:
    print("x is greater than 5")
elif x > 0:
    print("x is greater than 0 but not greater than 5")
else:
    print("x is less than or equal to 0")
```

In this example, since x has a value of 3, it is not greater than 5. The first if statement evaluates to false, so the program moves to the elif statement and checks the next condition x > 0, which is true. Therefore, the corresponding code block is executed, and the message "x is greater than 0 but not greater than 5" is printed.

Logical conditionals are powerful tools for decision-making in programming. They allow you to control the flow of your code and perform different actions based on different conditions. By combining if, elif, and else statements, you can create complex decision structures to handle a wide range of scenarios.

# *Loops*

Loops are like what it sounds. It allows you to loop blocks of code a certain number of times.

Python provides two types of loops: for loops and while loops.

## for Loop

A for loop is used to iterate over a sequence (such as a list, tuple, string, or range) or any other iterable object. It allows you to execute a block of code a predetermined number of times, once for each item in the sequence. The basic syntax of a for loop is as follows:

```python
'''
for item in sequence:
    Code to be executed for each item
'''

fruits = ['apple', 'banana', 'cherry']

for fruit in fruits:
    print(fruit)
```

In this example, the for loop iterates over each item in the fruits list. For each iteration, the current item is assigned to the variable fruit, and the corresponding block of code is executed. In this case, the code simply prints each fruit on a separate line.

You can also use the range() function to generate a sequence of numbers to iterate over. Here's an example:

```python
for number in range(1, 6):
    print(number)
```

In this example, the range(1, 6) function generates a sequence of numbers from 1 to 5 (6 is exclusive). The for loop iterates over each number in the range, and the code block inside the loop prints each number.

## while Loop

A while loop is used to repeatedly execute a block of code as long as a given condition is true. It allows you to keep looping until the condition becomes false. The basic syntax of a while loop is as follows:

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

In this example, the while loop executes the code block as long as the condition count <= 5 is true. The variable count starts with a value of 1, and within each iteration, it is incremented by 1 (count += 1). The loop continues until count becomes 6, at which point the condition becomes false, and the loop terminates.

## Loop Control Statements

Python provides several loop control statements that allow you to modify the behavior of loops:

  break statement: Terminates the loop prematurely, even if the loop condition is still true. It is used to exit the loop when a certain condition is met.

```python
fruits = ['apple', 'banana', 'cherry']

for fruit in fruits:
    if fruit == 'banana':
        break
    print(fruit)
```

In this example, when the loop encounters the fruit 'banana', the break statement is executed, and the loop is terminated.

  continue statement: Skips the remaining code in the current iteration of the loop and moves to the next iteration. It is used to skip specific items or conditions within a loop.

```python
fruits = ['apple', 'banana', 'cherry']

for fruit in fruits:
    if fruit == 'banana':
        continue
    print(fruit)
```

In this example, when the loop encounters the fruit 'banana', the continue statement is executed, and the current iteration is skipped. The loop then moves to the next iteration and continues with the remaining items.

Looping is a powerful concept in programming, and it allows you to perform repetitive tasks efficiently. ## **Final Coding Challenge: Mystical Math Marathon**
Steps:
  1. Import the math module to gain access to mathematical functions.

  2. Write a function named is_prime() that takes an integer as input and returns True if the number is prime and False otherwise. You can use this function to check if a given number is prime.

  3. Implement a function named find_mystic_number() that finds the Mystic Number. This function should iterate over a range of numbers and check if each number is prime using the is_prime() function. Once two prime numbers are found, multiply them together to obtain the Mystic Number.

  4. Finally, print the Mystic Number and celebrate your triumph!

Here's a starting point for your code:

```python
import math

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(math.sqrt(num)) + 1): # int(math.sqrt(num)) is to truncate the number if it is a decimal ex: 1.9 would be treated as 1.
        if num % i == 0:
            return False
    return True

def find_mystic_number():
    for i in range(2, 1000):
        if is_prime(i):
            for j in range(i + 1, 1000):
                if is_prime(j):
                    mystic_number = i * j
                    return mystic_number

# Call the function to find the Mystic Number
mystic_number = find_mystic_number()
print("The Mystic Number is:", mystic_number)
```
